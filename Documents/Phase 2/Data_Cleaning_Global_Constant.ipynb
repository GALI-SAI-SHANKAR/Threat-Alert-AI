Data cleansing or data cleaning is the process of detecting and correcting (or removing) corrupt or inaccurate records from a record set, table, or database and refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying, or deleting the dirty or coarse data.

Use a global constant to fill in the missing value: Replace all missing attribute values by the same constant, such as a label like “Unknown” or-. If missing values are replaced by, say, “Unknown,” then the mining program may mistakenly think that they form an interesting concept, since they all have a value in common-that of

“Unknown.” Hence, although this method is simple, it is not recommended.
